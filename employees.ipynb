Data Cleaning
Missing Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())
```

```python
df["Performance"] = df["Performance"].fillna(df["Performance"].median())
```

Feature Engineering


```python
df["Total income"] = df["Salary"] + df["Bonus"]
```

```python
def exp_level(x):
    if x <= 2:
        return "Beginner"
    elif 3 <= x <= 5:
        return "Intermediate"
    else:
        return "Senior"

df["ExperienceLevel"] = df["Experience"].apply(exp_level)
```

```python
def pef_grade(y):
    if y>=4.8:
        return "Outstanding"
    elif y>=4.5:
        return "Excellent"
    elif y>=4.0:
        return "Good"
    else:
        return "Average"
df["PerformanceGrade"] = df["Performance"].apply(pef_grade)
```

Visualization


```python
salary = df["Salary"]

plt.Figure(figsize=(8,5))
plt.hist(salary)
plt.title("Salary Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Frequency")
plt.ylabel("Salary")
plt.show()
```

```python
performance = df["Performance"]

plt.figure(figsize=(8,5))
plt.hist(performance, bins=10, color="skyblue", edgecolor="black")

plt.title("Performance Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Performance")
plt.ylabel("Frequency")
plt.show()
```

```python
skill = df["SkillScore"]

plt.figure(figsize=(8,5))
plt.hist(skill, bins=10, color="mediumseagreen", edgecolor="black", alpha=0.8)

plt.title("SkillScore Distribution", fontsize=14, fontweight="bold")
plt.xlabel("SkillScore")
plt.ylabel("Frequency")
plt.show()
```

```python
attendance = df["Attendance"]

plt.figure(figsize=(8,5))
plt.hist(attendance, bins=10, color="royalblue", edgecolor="white", alpha=0.9)
plt.title("Attendance Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
